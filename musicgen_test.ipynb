# MusicGen Test — Chinese Ambient Music
Testing if open-source models can match Flow Music quality for Silk Sound Dynasty.

```
# Install dependencies
!pip install -q audiocraft>=1.2.0 soundfile
```

```
from audiocraft.models import MusicGen
import soundfile as sf

# Load the best model
model = MusicGen.get_pretrained('facebook/musicgen-large')
model.set_generation_params(duration=30)  # 30 second sample
```

```
# Generate Chinese ambient music
prompts = [
    "Traditional Chinese guzheng and bamboo flute ambient music, peaceful meditation, gentle flowing melody, no drums, no vocals, zen garden atmosphere",
    "Erhu and guqin texture, Chinese classical ambient, slow tempo, water sounds, bamboo forest, calming spa music",
]

for i, prompt in enumerate(prompts):
    print(f"Generating: {prompt}")
    wav = model.generate([prompt], progress=True)
    sf.write(f'chinese_ambient_{i}.wav', wav[0].cpu().numpy().T, 32000)
    print(f"Saved chinese_ambient_{i}.wav")

from IPython.display import Audio
for i in range(len(prompts)):
    display(Audio(f'chinese_ambient_{i}.wav'))
```
