```
import torch
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import soundfile as sf
import IPython.display as ipd

processor = AutoProcessor.from_pretrained("facebook/musicgen-large")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-large")
model.to("cuda")
model.generation_config.max_new_tokens = 600  # ~30 seconds

print("Model loaded! Generating track 1...")

# Track 1: Chinese instrumental
prompt1 = "ancient chinese instrumental music, guzheng and erhu duet, slow tempo, natural dynamics, expressive human-like performance, traditional chinese scale, soft bamboo flute accents, minimal reverb, cinematic and emotional, peaceful atmosphere"
inputs = processor(text=[prompt1], padding=True, return_tensors="pt").to("cuda")
audio_values = model.generate(**inputs, do_sample=True, temperature=0.9)
sf.write('track1_chinese.wav', audio_values[0].cpu().numpy().T, 32000)
print('Track 1 done!')
ipd.display(ipd.Audio('track1_chinese.wav'))

# Track 2: Meditation
prompt2 = "meditation music with soft ambient pads and gentle flowing textures, slow tempo, deep calming atmosphere, minimal composition, no percussion, peaceful and immersive, designed for mindfulness and inner calm"
inputs = processor(text=[prompt2], padding=True, return_tensors="pt").to("cuda")
audio_values = model.generate(**inputs, do_sample=True, temperature=0.9)
sf.write('track2_meditation.wav', audio_values[0].cpu().numpy().T, 32000)
print('Track 2 done!')
ipd.display(ipd.Audio('track2_meditation.wav'))
```
