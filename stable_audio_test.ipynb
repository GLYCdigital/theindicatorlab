```
import torch
import soundfile as sf
import IPython.display as ipd
from diffusers import StableAudioPipeline

pipe = StableAudioPipeline.from_pretrained("stabilityai/stable-audio-open-1.0", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
print("Model loaded!")
```

```
# Track 1: Chinese instrumental
prompt1 = "ancient chinese instrumental music, guzheng and erhu duet, slow tempo, natural dynamics, expressive human-like performance, traditional chinese scale, soft bamboo flute accents, minimal reverb, cinematic and emotional, peaceful atmosphere"

audio1 = pipe(
    prompt1,
    num_inference_steps=200,
    audio_end_in_s=30,
    num_waveforms_per_prompt=1,
).audios[0]

sf.write('track1_chinese_stable.wav', audio1.T, pipe.vae.sampling_rate)
print('Track 1 done!')
ipd.display(ipd.Audio('track1_chinese_stable.wav'))
```

```
# Track 2: Meditation ambient
prompt2 = "meditation music with soft ambient pads and gentle flowing textures, slow tempo, deep calming atmosphere, minimal composition, no percussion, peaceful and immersive, designed for mindfulness and inner calm"

audio2 = pipe(
    prompt2,
    num_inference_steps=200,
    audio_end_in_s=30,
    num_waveforms_per_prompt=1,
).audios[0]

sf.write('track2_meditation_stable.wav', audio2.T, pipe.vae.sampling_rate)
print('Track 2 done!')
ipd.display(ipd.Audio('track2_meditation_stable.wav'))
```
