```
# Stable Audio Test - using transformers (pre-installed)
import torch
import soundfile as sf
import IPython.display as ipd
from transformers import pipeline

pipe = pipeline("text-to-audio", model="stabilityai/stable-audio-open-1.0", device=0)
print("Model loaded!")
```

```
# Track 1: Chinese instrumental
prompt1 = "ancient chinese instrumental music, guzheng and erhu duet, slow tempo, natural dynamics, expressive human-like performance, traditional chinese scale, soft bamboo flute accents, minimal reverb, cinematic and emotional, peaceful atmosphere"
result = pipe(prompt1, forward_params={"max_new_tokens": 480})
sf.write('track1_stable.wav', result["audio"][0].T, result["sampling_rate"][0])
print('Track 1 done!')
ipd.display(ipd.Audio('track1_stable.wav'))
```

```
# Track 2: Meditation ambient
prompt2 = "meditation music with soft ambient pads and gentle flowing textures, slow tempo, deep calming atmosphere, minimal composition, no percussion, peaceful and immersive, designed for mindfulness and inner calm"
result = pipe(prompt2, forward_params={"max_new_tokens": 480})
sf.write('track2_stable.wav', result["audio"][0].T, result["sampling_rate"][0])
print('Track 2 done!')
ipd.display(ipd.Audio('track2_stable.wav'))
```
